### Building the Machine Learning Model

```python
# dividing our data into Feature and Target

X= df.drop('crop_yield', axis = 1)
y = df['crop_yield']
```

### Importing the model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import  r2_score, mean_squared_error, mean_absolute_error
```

```python
# train the model

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .3, random_state=42)
```

```python
# Normalize the data

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
# initialize the model algorithm

lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
lr
```

```python
lr.coef_
```

```python
# test the model by predicting

y_pred = lr.predict(X_test_scaled)

y_pred
```

```python
# coefficient of determination

R2 = r2_score(y_test, y_pred)

R2
```

```python
# Accuracy Score
print('R Square:', R2)
print('Mean Square Error:', mean_squared_error(y_test, y_pred))
print('Mean Absolute Error:',mean_absolute_error(y_test, y_pred))
```

## Final Result

### 1 The farm_size determines the crop_yield as evident in the correlation, Spearman's correlation of 0.99, other factors like fertilizer, sunlight duration, rainfall, and soil_quality have little effect on the crop yield.

### 2 The model is perfect as shown by the R squared of 0.99 which shows the model is able to explain 99% of variation in crop yield, MSE of 0.08 explain the model error is minimal and MAE of 0.24 explain the model predictions are off by about 0.24 units.
