### Data Cleaning

```python
# cleanng extra space in department column

df['department'] = df.department.str.strip()
# correcting sweing to sewing
df['department'] = df.department.replace(to_replace ='sweing', value='sewing')

df['department'].value_counts()
```

#### removing date column
####  removing idle_time and idle_men due to few non-zeros values
####  removing wip due to many null values
####  removing no_of_style_change due to fe non-zero values

```python
df_n = df.drop(['date','wip','idle_time', 'idle_men','no_of_style_change'], axis=1)

df_n.head(5)
```

```python
# merging Quarter 5 with Quarter 4 because it covers so few values as its num

df_n.loc[df_n['quarter'] == 'Quarter5', 'quarter']= 'Quarter4'

df_n.quarter.value_counts()
```

```python
# making quarter numeric

quarter_map = {'Quarter1':1, 'Quarter2':2, 'Quarter3':3, 'Quarter4':4}

df_n.quarter = df_n['quarter'].replace(quarter_map).astype('int64')

df_n.quarter.value_counts()
```

```python
# making number_of_workers column integer instead of float

df_n['no_of_workers'] = df_n.no_of_workers.astype('int64')
print(df_n['no_of_workers'].dtype)
df_n.head(5)
```

```python
# actual_productivity should be of the same decimal places with targeted_productivity

df_n['actual_productivity'] = df_n.actual_productivity.round(2)

df_n.head(5)
```

```python
df_n['productive'] = df_n.actual_productivity >= df_n.targeted_productivity

df_n.sample(10, random_state=14)
```

Machine Learning Preparation

```python
# covert department column to boolean

df_n = df_n.rename(columns= {'department':'dept_sewing'})
df_n['dept_sewing'] = df_n['dept_sewing'].map({'finishing': 0, 'sewing': 1})

df_n.head(5)
```

```python
# make quarter columns into dummies

dummy = pd.get_dummies(df_n['quarter'], prefix='q')

# concat df_n with dummy

df_n = pd.concat([df_n,dummy], axis=1)



df_n.head(5)
```

```python
df_n = df_n.drop('quarter', axis=1)
```

```python
df_n.head(5)
```

```python
# make day columns into dummies

dummy2 = pd.get_dummies(df_n['day'])

# concat df_n with dummy

df_n = pd.concat([df_n,dummy2], axis=1)



df_n.head(5)
```

```python
df_n = df_n.drop('day', axis=1)
```

```python
# make team columns into dummies

dummy3 = pd.get_dummies(df_n['team'], prefix='team')

# concat df_n with dummy

df_n = pd.concat([df_n,dummy3], axis=1)

df_n = df_n.drop('team', axis=1)

df_n.head(5)
```

### Building the tree

```python
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
# feature and target columns

X = df_n.drop(['actual_productivity','productive'], axis=1)
y = df_n['productive']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle=True, random_state=24)
```

```python
tree = DecisionTreeClassifier(max_depth=3, random_state=24)

tree.fit(X_train, y_train)
```

```python
y_pred = tree.predict(X_test)
```

### Visualizing and Evaluating the Tree

```python
from sklearn.metrics import accuracy_score

print('Accuracy:',round(accuracy_score(y_test, y_pred),2))
```

```python
#print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}%")
```

```python
tree.classes_
```

```python
plt.figure(figsize=(18,9))

plot_tree(tree,
          filled = True,
          rounded = True,
          feature_names = X.columns,
          class_names = ['Unproductive','Productive'],
          fontsize = 11)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

confusion_matrix(y_test, y_pred)
print('Precision_score:',round(precision_score(y_test, y_pred),2))
print('Recall_score:',round(recall_score(y_test, y_pred),2))
print('F1_score:',round(f1_score(y_test, y_pred),2))
print('Accuracy:',round(accuracy_score(y_test, y_pred),2))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree, X, y, cv=10)

print('Cross Validation Accuracy Scores:', scores.round(2))
print('Mean Cross Validation Accuracy Scores:', scores.mean().round(2))
```
